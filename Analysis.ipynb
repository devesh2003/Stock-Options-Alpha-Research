```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
options = pd.read_csv("options-data-more.csv")
prices = pd.read_csv("Oct-Dec.csv")
# prices.drop([18,19,20,21],axis=0,inplace=True)
prices["Change"] = ((prices["Close Price"] - prices["Open Price"])/prices["Open Price"])*100
options["Change"] = ((options["Close"] - options["Open"])/options["Open"])*100
len(prices["Date"])
prices.head()
```

```python
# xs = options["Date"]
xs = np.array([x for x in range(1,len(prices["Date"]) + 1)])
ys = prices["Change"]
y2 = options["Change"]
```

```python
y3 = options["Change"] - prices["Change"]
plt.bar(xs,y3)
plt.title("Difference Between Options & Stock % Change ")
# plt.xticks(xs,options["Date"])
```

```python
y4 = options["Change"]
plt.bar(xs,y4)
plt.title("% Change in Option Price")
```

```python
y5 = prices["No. of Trades"] / prices["Change"].abs()
plt.plot(xs,y5)
plt.title("Ratio of Trades and % Change")
```

```python
y6 = prices["Total Traded Quantity"] / prices["No. of Trades"]
plt.plot(xs,y6)
plt.title("Average Quantity Per Trade")
```
