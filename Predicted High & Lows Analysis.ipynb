```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = pd.read_csv("banknifty-prices.csv")
```

```python
prices.head()
```

```python
val = 0.75

prices["predhigh"] = (prices["Close"].shift(1))*(100+val)/100
prices["predlow"] = (prices["Close"].shift(1))*(100-val)/100
```

```python
prices.head()
```

```python
prices.loc[prices.predhigh <= prices.High, ["sell"]] = 1
prices.loc[prices.predhigh > prices.High, ["sell"]] = 0
prices.loc[prices.predlow >= prices.Low, ["buy"]] = 1
prices.loc[prices.predlow < prices.Low, ["buy"]] = 0
```

```python
prices.loc[prices.predlow >= prices.Low, ["final_buy"]] = prices.predlow
prices.loc[prices.predhigh <= prices.High, ["final_sell"]] = prices.predhigh
```

```python
prices.loc[((prices.sell==1)&(prices.buy==0)),'final_buy'] = prices.Close
prices.loc[((prices.sell==0)&(prices.buy==1)),'final_sell'] = prices.Close
```

```python
prices.head()
```

```python
prices['profit_buy'] = 100*(prices.final_buy-prices.Close)/prices.final_buy
```
